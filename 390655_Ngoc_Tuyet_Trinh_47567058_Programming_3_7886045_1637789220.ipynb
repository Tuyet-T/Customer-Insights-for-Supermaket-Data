```python
import pandas as pd
df_legend = pd.read_csv('data/data legend.csv')
df = pd.read_csv('data/data.csv')

#PART I: DATA EXPLORATORY
info = df.info()
info
stats = df.describe()
print(stats)

#Data visualization
import matplotlib.pyplot as plt

#Pie chart for sex
label = {0: "male", 1: "female"}
Sex_count = df['Sex'].value_counts()
Sex_df = pd.DataFrame({'label': Sex_count.index, 'Count': Sex_count}) 
Sex_df['label'] = Sex_df['label'].map(label)
plt.pie(Sex_df['Count'], labels=Sex_df['label'], autopct='%1.1f%%',colors=['limegreen','red'],startangle=90)
plt.title("Sex Distribution")
plt.axis('equal')
plt.show()

#Pie chart for Maritual status
label = {0: "single", 1: "non-single"}
Ma_count = df['Marital status'].value_counts()
Ma_df = pd.DataFrame({'label': Ma_count.index, 'Count': Ma_count}) 
Ma_df['label'] = Ma_df['label'].map(label)
plt.pie(Ma_df['Count'], labels=Ma_df['label'], autopct='%1.1f%%',colors=['orange','yellow'],startangle=90)
plt.title("Marital status Distribution")
plt.axis('equal')
plt.show()

#Pie chart for Settlement size
label = {0: "small city", 1: "mid_sized", 2: "big city"}
settlement_count = df['Settlement size'].value_counts()
settlement_df = pd.DataFrame({'label': settlement_count.index, 'Count': settlement_count}) 
settlement_df['label'] = settlement_df['label'].map(label)
plt.pie(settlement_df['Count'], labels=settlement_df['label'], autopct='%1.1f%%',colors=['limegreen','red','orange'],startangle=90)
plt.title("Settlement size Distribution")
plt.axis('equal')
plt.show()

#Pie chart for Occupation
label = {0: "unemployed/unskilled", 1: "skilled people/official", 2: "management / self-employed / highly qualified employee / officer"}
occupation_count = df['Occupation'].value_counts()
occupation_df = pd.DataFrame({'label':occupation_count.index, 'Count': occupation_count}) 
occupation_df['label'] = occupation_df['label'].map(label)
plt.pie(occupation_df['Count'], labels=occupation_df['label'], autopct='%1.1f%%',colors=['limegreen','red','orange'],startangle=30)
plt.title("Occupation Distribution")
plt.axis('equal')
plt.show()

#Pie chart for Education
label = {0: "unknown", 1: "high school", 2: "university", 3: "graduate school"}
edu_count = df['Education'].value_counts()
edu_df = pd.DataFrame({'label': edu_count.index, 'Count': edu_count}) 
edu_df['label'] = edu_df['label'].map(label)
plt.pie(edu_df['Count'], labels=edu_df['label'], autopct='%1.1f%%',colors=['limegreen','yellow','red','orange'])
plt.title("Education level distribution")
plt.axis('equal')
plt.show()

#Scatter plot for Income and Age
x = df['Age']
y = df['Income']
plt.scatter(x, y,c='limegreen')
plt.xlabel("Age")
plt.ylabel("Annual Income in USD")
plt.title('Income and Age distribution') 
plt.show()

#PART 2
#Optimal cluster number by Elbow method
from sklearn.cluster import KMeans
import warnings 
warnings.filterwarnings("ignore")

inertias = [] 
feature = df.drop(['ID'], axis=1)
for i in range(1, 11):
    km = KMeans(n_clusters=i, 
                init='k-means++', 
                n_init=10, 
                max_iter=300, 
                random_state=0)
    km.fit(feature)
    inertias.append(km.inertia_)

plt.plot(range(1, 11), inertias, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Cluster inertia (within-cluster SSE)')
plt.xticks(range(1,11))
plt.title('Elbow method') 
plt.tight_layout()
plt.show()

#Kmeans method to estimate the cluster
Kmean_plus = KMeans(n_clusters=4, 
            init='k-means++',
            n_init = 10, 
            max_iter=300,
            random_state=0)

y_kmean_plus = Kmean_plus.fit_predict(feature)

#print(y_kmean_plus)
print(f'\nClusters:\n {Kmean_plus.cluster_centers_}')

#Hierarchical clustering
df2= df.drop(['ID'], axis=1)
from scipy.spatial.distance import pdist, squareform
import numpy as np

row_dist = pd.DataFrame(squareform(pdist(df2, metric='euclidean')))

pd.Series(pdist(df2, metric='euclidean'))

from scipy.cluster.hierarchy import linkage


row_clusters = linkage(pdist(df2, metric='euclidean'), method='complete')
pd.DataFrame(row_clusters,
             columns=['row label 1', 'row label 2',
                      'distance', 'no. of items in clust.'],
             index=[f'cluster {i + 1}' for i in range(row_clusters.shape[0])])

from scipy.cluster.hierarchy import dendrogram

from scipy.cluster.hierarchy import set_link_color_palette
set_link_color_palette(['limegreen'])

row_dendr = dendrogram(row_clusters, 
                       color_threshold=np.inf
                       )
plt.tight_layout()
plt.ylabel('Euclidean distance')
plt.title('Hierarchical clustering')
plt.show()

from scipy.cluster.hierarchy import cut_tree
Customer_cluster = cut_tree(row_clusters, 4).reshape(-1,)
df2['Customer_cluster'] = Customer_cluster
Cluster_group = df2[['Sex','Marital status','Age','Education','Income','Occupation','Settlement size','Customer_cluster']].groupby('Customer_cluster').mean()
Cluster_group
```
